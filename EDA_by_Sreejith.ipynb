```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/Car_sales.csv")
df
```

```python
print(df.describe())
```

```python
print(df['Vehicle_type'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
bins = [0, 50, 100, 150, 200, 300, 600]
sales_counts = pd.cut(df['Sales_in_thousands'], bins=bins, include_lowest=True, right=False).value_counts().sort_index()
plt.bar(sales_counts.index.astype(str), sales_counts.values, color='#4CAF50', edgecolor='#388E3C')
plt.title('Sales Distribution')
plt.xlabel('Sales (in thousands)')
plt.ylabel('Number of Models')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10,6))
bins = [0, 20, 40, 60, 80, 100, 150]
price_counts = pd.cut(df['Price_in_thousands'], bins=bins, include_lowest=True, right=False).value_counts().sort_index()
plt.bar(price_counts.index.astype(str), price_counts.values, color='#2196F3', edgecolor='#1565C0')
plt.title("Price Distribution")
plt.xlabel("Price (in thousands)")
plt.ylabel("Number of Models")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
bins = [0, 100, 150, 200, 250, 300, 400, 600]
hp_price = df.groupby(pd.cut(df['Horsepower'], bins=bins, include_lowest=True, right=False))['Price_in_thousands'].mean()
plt.bar(hp_price.index.astype(str), hp_price.values, color='#9C27B0', edgecolor='#6A1B9A')
plt.title("Average Price by Horsepower Range")
plt.xlabel("Horsepower Range")
plt.ylabel("Average Price (in thousands)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Engine_size'], df['Price_in_thousands'], alpha=0.6, color='red', edgecolor="black")
plt.title("Engine Size vs Price")
plt.xlabel("Engine Size")
plt.ylabel("Price (in thousands)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_manu = df.groupby('Manufacturer')['Sales_in_thousands'].sum().sort_values(ascending=False).head(10)
plt.bar(top_manu.index, top_manu.values, color='#009688', edgecolor='#004D40')
plt.title("Top 10 Manufacturers by Sales")
plt.xlabel("Manufacturer")
plt.ylabel("Total Sales (in thousands)")
plt.xticks(rotation=45)
plt.show()
```
